```python
import numpy as np 
from provabgs import models as Models
```

```python
# --- plotting --- 
import corner as DFM
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
```

```python
m_nmf = Models.NMF(burst=True, iso_unc=True, emulator=True)
```

```python
# some arbitrary SPS parameter values with no **isochrone offset**
theta0 = np.array([10., 1., 0., 0., 0., 0., 10., 1e-3, 1e-3, 1., 1., 0., 0., 0., 0.])
```

```python
fig = plt.figure(figsize=(10,5))
sub = fig.add_subplot(111)

w, f = m_nmf.sed(theta0, 0.1)
sub.plot(w, f, c='k')

sub.set_xlim(3e3, 1e4)
```

```python
fig = plt.figure(figsize=(10,15))
sub = fig.add_subplot(311)

w, f = m_nmf.sed(theta0, 0.1)
sub.plot(w, f, c='k')
for i in range(3): 
    _theta = np.array([10., 1., 0., 0., 0., 0., 10., 1e-3, 1e-3, 1., 1., 0., 1.+float(i), 0., 0.])
    w, f = m_nmf.sed(_theta, 0.1)
    sub.plot(w, f)

sub.set_xlim(3e3, 1e4)

sub = fig.add_subplot(312)

w, f = m_nmf.sed(theta0, 0.1)
sub.plot(w, f, c='k')
for i in range(3): 
    _theta = np.array([10., 1., 0., 0., 0., 0., 10., 1e-3, 1e-3, 1., 1., 0., 0., 1.+float(i), 0.])
    w, f = m_nmf.sed(_theta, 0.1)
    sub.plot(w, f)

sub.set_xlim(3e3, 1e4)


sub = fig.add_subplot(313)

w, f = m_nmf.sed(theta0, 0.1)
sub.plot(w, f, c='k')
for i in range(3): 
    _theta = np.array([10., 1., 0., 0., 0., 0., 10., 1e-3, 1e-3, 1., 1., 0., 0., 0., 1.+float(i)])
    w, f = m_nmf.sed(_theta, 0.1)
    sub.plot(w, f)

sub.set_xlim(3e3, 1e4)
```
