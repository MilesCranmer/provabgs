```python
import numpy as np
from provabgs import models as Models
import matplotlib.pyplot as plt
```

```python
fsps = Models.FSPS(name='nmf_bases')
```

```python
fig = plt.figure(figsize=(10, 5))
sub = fig.add_subplot(111)
for i in range(4): 
    theta = np.array([0., 0., 0., 0., 1e-4, 1e-4, 0, 0, 0, 0.1])
    theta[i] = 1
    w, lum = fsps._fsps_nmf(theta)
    theta[-1] = fsps._tage_z_interp(0.1)
    w0, lum0 = fsps._fsps_nmf_v0(theta) 
    
    print(np.abs(lum - lum0).max())
    
    sub.plot(w, lum, c='k')
    sub.plot(w0, lum0, c='C1')
    sub.set_xlim(3e3, 1e4)
```
